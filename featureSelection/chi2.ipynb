```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
df=pd.read_csv('/content/data.csv')
```

```python
def chi2_feature_selection(dataFrame,targetColumn):
  p_values=[]
  for column in dataFrame.columns:
    if column != targetColumn:
      ct=pd.crosstab(df[column],df[targetColumn])
      _,p_value,_,_=chi2_contingency(ct)
      p_values.append((column,p_value))
  return p_values
```

```python
result=chi2_feature_selection(df,'Target')
```

```python
result_df=pd.DataFrame(result,columns=['Feature','p-value']).sort_values(by='p-value')
```
